**GET request**

Make a GET Request to Metaweather website. GET Request is made to request data from the website

https://www.metaweather.com/api/

location : Sanfransico

/api/location/(woeid)/

Date in the format yyyy/mm/dd.

```python
r_get = requests.get('https://www.metaweather.com/api/location/2487956/2019/07/02')
```

Request is successful when the status code is 200

```python
r_get.status_code
```

Check the type of req variable. It is a Response object.

```python
type(r_get)
```

```python
type(r_get.headers)
```

```python
pprint(r_get.headers)
```

```python
data = json.loads(r_get.text)

pprint(data)
```
