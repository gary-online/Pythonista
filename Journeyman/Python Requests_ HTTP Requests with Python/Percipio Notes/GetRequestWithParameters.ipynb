**Retrieving the search keyword from the user**

The code takes a search query from the user and passes as parameter to the Youtube url.

```python
search_term = input("Enter the term you need to search: ")
```

#### The request will be made to youtube.com/search

```python
URL = 'https://www.youtube.com/search'
```

**The parameters can be passed as a dictionary**

```python
PARAMS = {'q' : search_term}
```

```python
r_get = requests.get(url = URL,
                     params = PARAMS)
```

```python
print(r_get.status_code)
```

```python
print(r_get.url)
```

**Enter the above URL in your browser to verify it.**

```python
webbrowser.open(r_get.url)
```
