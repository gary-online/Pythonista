# The Graph

## Overview
1. A graph is used to represent relationships between entities
2. The entities can be anything - graphs find applications in variety of ways in the real world
3. These relationships can be arbitrarily complicated and a variety of different types
4. Graphs are used to represent information in many real world applications
5. Graphs are also favourite interview questions - they can start from simple concepts and get arbitrarily complex
6. Graph theory is a complex field of study by itself - there are many algorithms to optimize different problems represented using graphs
7. We'll only be scratching the surface here - however this should give a strong foundation to solve graph problems from first principles

## What is a graph?
1. A set of vertices and edges
2. Two vertices and a single edge is also a valid graph

### Undirected
Undirected edges represent 2 way relationships such as
1. Two-way roads
2. I am his friend and he is mine

### Directed
Directed edges represent 1-way relationships such as:
1. One-way roads
2. I report to my manager

## What makes up a graph?
1. Vertex
2. Edges

### Vertex
1. The entities are people
2. The entities are locations
3. Entities are computers across the world

### Edge
1. Professional graph - professional relationships / people that work together; Social graph, as in a social network - Facebook 
2. Maps - a way to get from point a to b; each of these can be thought of by means of transport
3. The internet - a way to send information or data from one computer to another, info can be routed wireless or over wires


## What do you need to represent in a graph?
1. A way to model a vertex which may hold some information
2. A way to model Directed or Undirected EDGES

### Adjacency Matrix
1. Use a matrix with rows and columns
2. A matrix is just a table
3. The row labels and the column labels represent the vertices
4. Each cell represents the relationship between the vertices i.e. the edges
5. A graph with n vertices has n^2 cells in its adjacency matrix
6. The complexity of the operation to look up whether an edge is present between two vertices in a graph when it is represented as an adjacency matrix is O(1)


## Graphs as an Adjacency List and Set

### Adjacency List:
1. Each vertex is a node
2. Each vertex has a pointer to a linked list
3. This linked list contains all the other nodes this vertex connects to directly
4. Adjacency lists are more space-efficient than adjacency matrices
5. Adjacency lists are not the best representations of graphs
6. The order of the vertices in the adjacency list does not matter
7. The same graph can have multiple representations
8. Certain operations become tricky e.g. deleting a node involves looking through all the adjacency lists to remove the node from all lists.


### Adjacency Set
1. It is similar to an adjacency list - except that the list of adjacent vertices for each node is a set rather than a linked list


Adjacency Matrix works well when the graph is connected i.e. many nodes are connected to many other nodes. The overhead of V2 space is worth it when the number of connections are large.

Adjacency List and Adjacency Set: A sparse graph with few connections between nodes might be more efficiently represented using adjacency list of a set
